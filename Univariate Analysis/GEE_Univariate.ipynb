## BOXPLOT

```python
state = data.copy()

for st in np.unique(state['STATE_UT']):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    state_n = state[state['STATE_UT'] == st]
    bplot1 = ax1.boxplot(x="maximum_2m_air_temperature", data=state_n, patch_artist=True)
    ax1.set_title('MAXIMUM')
    ax1.set_xlabel(str(st))

    bplot2 = ax2.boxplot(x="minimum_2m_air_temperature", data=state_n, patch_artist=True)
    ax2.set_title('MINIMUM')
    ax2.set_xlabel(str(st))
    
    colors = ['gold']
    for bplot in (bplot1, bplot2):
        for patch, color in zip(bplot['boxes'], colors):
            patch.set_facecolor(color)
    plt.show()
```

### A box plot (also known as box and whisker plot) is a type of chart often used in explanatory data analysis to visually show the distribution of numerical data and skewness through displaying the data quartiles (or percentiles) and averages.
### Box plots show the five-number summary of a set of data: including the minimum score, first (lower) quartile, median, third (upper) quartile, and maximum score.
### They are used for :
- ### They show the average score of a data set.
- ### They show the skewness of a data set.
- ### They show the dispersion of a data set.
- ### They show outliers within a data set.



## LAG PLOT

> ### Lag plots are most commonly used to look for patterns in time series data.

```python
state = data.copy()

for st in np.unique(state['STATE_UT']):
    state_n = state[state['STATE_UT'] == st]
    s = pd.Series(state_n['maximum_2m_air_temperature'])
    plt.title(str(st))
    pd.plotting.lag_plot(s, lag=1)
    plt.show()
```

## One set of observations in a time series is plotted (lagged) against a second, later set of data. The kth lag is the time period that happened k time points before time i.Here lag=1 and observation is maximum temperature of all states

## STRIP PLOT


```python
kkd = data[data["NAME"]=="KOZHIKODE"]
plt.subplots(figsize=(20,10))
sns.stripplot(x = kkd["date"].dt.month, y = kkd['minimum_2m_air_temperature'], hue=kkd["date"].dt.month)
plt.legend(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.title('Strip Plot for Kozhikode District', fontsize=20)
plt.xlabel('Month',fontsize=20)
plt.ylabel('Minimum Temperature',fontsize=20)
plt.show()
```

### Plot in cases where you want to show all observations along with some representation of the underlying distribution.
### A strip is basically a scatter plot where the x axis represents a categorical variable. Typical uses of a strip plot involves applying a small random jitter value to each data point such that the separation between points becomes clearer

>- ### **El Nio** is a climate pattern that describes the unusual warming of surface waters in the eastern tropical Pacific Ocean.
>- ### El Nino conditions arise from the episodic warming of the tropical Pacific off Peru and lead to surges of extreme drought and heat in the Northern and Southern hemispheres.
